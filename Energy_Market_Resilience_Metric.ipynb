```python
#EDA - Investigating the categorical columns
#specifically, we will examine:
#location/Region and Energy Source from historical data
#Infrastructure status, maintenance activities, and technology limitations from infrastructure data
#Competitor Data and market trends from market data
#Regulatory changes and compliance status from regulatory data

#grouping all datasets as dataset
datasets ={"historical":historical,
    "infrastructure":infrastructure,
    "market":market,
    "regulatory":regulatory    
}


#Extracting unique values on their counts for categorical columns

categorical_columns = {
    "historical":["Location/Region","Energy Source"],
    "infrastructure":["Maintenance Activities","Technology Limitations"],
    "market":["Competitor Data","Market Trends"],
    "regulatory":["Regulatory Changes","Compliance Status"]
}

categorical_values_counts ={}
for dataset, columns in categorical_columns.items():
    for column in columns:
        categorical_values_counts[(dataset, column)] = datasets[dataset][column].value_counts()
                                      
```

```python
#visualizing 
#vcalculate the number of rows and columns for the subplots
num_rows = (len(categorical_values_counts) + 1) // 2
num_cols = 2

#create a grid of the subplots
fig, axes = plt.subplots(num_rows, num_cols, figsize = (12, 12))

#flatten the axes array for easier indexing
axes = axes.flatten()

#iterate through categorical columns and plot them in subplots
for i, ((dataset, column), counts) in enumerate(categorical_values_counts.items()):
    ax= axes[i]
    sns.barplot(x=counts.values, y=counts.index, ax=ax)
    ax.set_title(f"Value Counts for {dataset} - {column}")
    ax.set_xlabel("Count")
    ax.set_ylabel(column)

#hide any remaining subplots
for i in range(len(categorical_values_counts), num_rows * num_cols):
    fig.delaxes(axes[i])
    
    #adjust layout
    plt.tight_layout
    
    plt.show()
```

```python
#Energy demand, productiton and consumption over time
#Examine the trends in energy demand, production and consumption from the historicalenergy dataset
#This will give us insights into flunctuations and potential vulnerabilities in the energy market
```

```python
#Extract year and month from 'date/time' column
historical['Year'] = historical['Date/Time'].dt.year
historical['Month'] = historical['Date/Time'].dt.month

#create a new column 'Year-Month' for easy plotting
historical['Year-Month'] = historical['Date/Time'].dt.to_period('M')

#Aggregating data on a monthly basis
monthly_data = historical.groupby('Year-Month').mean()

#setting figsize
plt.figure(figsize=(15, 6))

#Plotting 
sns.lineplot(data=monthly_data, x=monthly_data.index.astype(str), y='Energy Production (kWh)', label='Energy Production', color= "blue", linestyle="-", linewidth=1.5, ci=None)
sns.lineplot(data=monthly_data, x=monthly_data.index.astype(str), y='Energy Consumption (kWh)', label='Energy Consumption', color= "red", linestyle="-", linewidth=1.5, ci=None)
sns.lineplot(data=monthly_data, x=monthly_data.index.astype(str), y='Energy Demand', label='Energy Demand', color= "green", linestyle="-", linewidth=1.5, ci=None)

#Setting titles and labels
plt.title('Monthly Aggregate of Energy Production, Consumption and Demand Over Time')
plt.xlabel('Date')
plt.ylabel('KWh')
plt.legend(loc="upper left", bbox_to_anchor=(1, 1))
plt.grid(True, which='both', linestyle='--', linewidth=0.5)

#Setting our labels
labels = monthly_data.index.astype(str).tolist()
n = 6
plt.xticks(labels[::n], rotation=360)

plt.tight_layout()
plt.subplots_adjust(hspace=0.5)

plt.show()
```

```python
#Energy Price and Market Price Over Time
    #To understand the pricingdynamics and its relation to market trends
    #Visualization of the market price from the market dataset along the Energy Price from the historical dataset over time
```

```python
#merging historical data and market data on Date/Time for combined analysis
merged_data = pd.merge(historical, market, on="Date/Time", how="inner")

#Setting 'Date/Time' as the index of the Dataframe for aggregation
merged_data['Year'] = merged_data['Date/Time'].dt.year
merged_data['Month'] = merged_data['Date/Time'].dt.month

#Aggregating data on monthly basis
monthly_merged_data = merged_data.groupby(['Year', 'Month']).mean()

#Create a new 'Year-Month' column for plotting
monthly_merged_data['Year-Month']= monthly_merged_data.index.get_level_values(0).astype(str)+ '-'+ monthly_merged_data.index.get_level_values(1).astype(str)

#setting figsize
plt.figure(figsize=(15, 6))

#Plotting 
sns.lineplot(data=monthly_merged_data, x='Year-Month', y='Market Price', label='Market Price', color= "blue", linestyle="-", linewidth=1.5, ci=None)
sns.lineplot(data=monthly_merged_data, x='Year-Month', y='Energy Price', label='Energy Price', color= "yellow", linestyle="-", linewidth=1.5, ci=None)

plt.title('Monthly Aggregate of Market Price and Energy Price Over Time')
plt.xlabel('Date')
plt.ylabel('Price ($)')
plt.legend(loc="upper left", bbox_to_anchor=(1, 1))
plt.grid(True, which='both', linestyle='--', linewidth=0.5)

#Setting our labels
labels = monthly_data.index.astype(str).tolist()
n = 6
plt.xticks(labels[::n], rotation=360)

plt.tight_layout()
plt.subplots_adjust(hspace=0.5)

plt.show()
```

```python
#Setting up the figures and axes
fig, axes = plt.subplots(1, 2, figsize=(15, 6))

#plotting the frequency of infrastructure data
sns.countplot(data=infrastructure, x='Infrastructure Status', ax=axes[0], order=['Good', 'Fair', 'Poor'],palette='viridis')
axes[0].set_title('Frequency of Infrastructure Status')
axes[0].set_xlabel('Infrastructure Status')
axes[0].set_ylabel('Count')

#Plotting the frequency of Technology Limitation
sns.countplot(data=infrastructure, x='Technology Limitations', ax=axes[1],palette='viridis')
axes[1].set_title('Frequency of Technology Limitations')
axes[1].set_xlabel('Technology Limitations')
axes[1].set_ylabel('Count')

plt.tight_layout()
plt.show()
```

```python
#Correlation between Infrastructure Status and Tech Limitations
#Cross tabulation
ct = pd.crosstab(infrastructure['Infrastructure Status'], infrastructure['Technology Limitations'])

plt.figure(figsize=(10, 6))
sns.heatmap(ct, annot=True,cmap='viridis')
plt.title('Infrastructure Status vs Technology Limitations')
plt.xlabel('Technology Limitations')
plt.ylabel('Infrastructure Status')

plt.tight_layout()

plt.show()
```

```python
#Correlation of High tech Limitations, Poor Infrastructure Status AND energy
#convert the columns to binary
infrastructure['Poor_Infrastructure'] = infrastructure['Infrastructure Status'].apply(lambda x: 1 if x == 'Poor' else 0)
infrastructure['High_Tech_Limitations'] = infrastructure['Technology Limitations'].apply(lambda x: 1 if x == 'High' else 0)

#combining columns
infrastructure['Combined']= infrastructure['Poor_Infrastructure'] * infrastructure['High_Tech_Limitations']

merged_data = pd.merge(historical, infrastructure, on="Date/Time", how="inner")

correlation_coefficient = merged_data['Energy Production (kWh)'].corr(merged_data['Combined'])
```

```python
#Renewables
#Cost and Revenue
#Impact of Regulatory Changes and Operational Costs on Revenue
# Visualize the relationship between Regulatory changes and compliance costs as well as the operational costs to the Revenue generated from the company.
#Revenue = Energy Consumption * Energy Price
# This will provide insights into thefinancial implications of regulatory changes socts and operational costs for Energix Enterprise.
```

```python
#calculate Revenue
historical['Revenue'] = historical['Energy Price'] * historical['Energy Consumption (kWh)']

#merge historical and regulatory data
merged_regulatory= pd.merge(historical, regulatory, on="Date/Time", how="inner") 

#Extract month and year
merged_regulatory['Year'] = merged_regulatory['Date/Time'].dt.year
merged_regulatory['Month'] = merged_regulatory['Date/Time'].dt.month

#Aggregate monthly data
monthly_aggregated_data=merged_regulatory.groupby(['Year', 'Month']).mean().reset_index()

monthly_aggregated_data['Year-Month']= monthly_aggregated_data['Year'].astype(str) + '-' + monthly_aggregated_data['Month'].astype(str)
plt.figure(figsize=(15, 8))

plt.fill_between(monthly_aggregated_data['Year-Month'], monthly_aggregated_data['Operational Costs'], color='blue', label='Operational Costs', alpha=0.5)
plt.fill_between(monthly_aggregated_data['Year-Month'], monthly_aggregated_data['Operational Costs'], monthly_aggregated_data['Operational Costs'] + monthly_aggregated_data ['Compliance Costs'],color='green', label='Compliance Costs', alpha=0.5)

                 
sns.lineplot(data=monthly_aggregated_data, x='Year-Month', y='Revenue', label= 'Revenue', color='red', linewidth=2)
                 
plt.title('Impact of Regulatory Changes on Operational, Revenue, and Compliance Costs (Monthly Aggregated)')
plt.xlabel('Date')
plt.ylabel('Amount ($)')
plt.legend(loc='upper left')
plt.grid(True, which='both', linestyle='--', linewidth=0.5)

#Adjust x-ticks
labels=monthly_aggregated_data['Year-Month'].tolist()
n = 4
plt.xticks(labels[::n], rotation=360)

plt.tight_layout()

plt.show()                 
```

```python
#Renewables
#Analyzing Competition from Renewable Energy Providers

#To understand the competition from renewable energy providers and its impact on Energix Enterprise. we will analyze the Energy Source column from the historical dataset.
# We'll visualize the trends in Energy productionn based on the Energy source (Fossil Fuels vs Renewables) over time. It will give insight into how the production from renewables has evolved and potentially affected the company's market share. 
```

```python
# Extract month and year
historical['Year'] = historical['Date/Time'].dt.year
historical['Month'] = historical['Date/Time'].dt.month

# Aggregate monthly data
monthly_aggregated_data = historical.groupby(['Year', 'Month', 'Energy Source']).sum()['Energy Production (kWh)'].reset_index()

# Create year-month column
monthly_aggregated_data['Year-Month'] = monthly_aggregated_data['Year'].astype(str) + '-' + monthly_aggregated_data['Month'].astype(str)

# Convert Year-Month to categorical
#monthly_aggregated_data['Year-Month'] = pd.Categorical(monthly_aggregated_data['Year-Month'], ordered=True)

# Set up the plot
plt.figure(figsize=(12, 4))

# Plot data for each energy source
for source in monthly_aggregated_data['Energy Source'].unique():
    source_data = monthly_aggregated_data[monthly_aggregated_data['Energy Source'] == source]
    plt.plot(source_data['Year-Month'], source_data['Energy Production (kWh)'], label=source)

plt.title('Monthly Aggregated Energy Production Trends by Energy Source Over Time')
plt.xlabel('Date')
plt.ylabel('Energy Production (kWh)')
plt.legend(title='Energy Source')
plt.grid(True, which='both', linestyle='--', linewidth=0.5)

# Adjust x-ticks
labels = monthly_aggregated_data['Year-Month'].tolist()
n = 4
#plt.xticks(labels[::n], rotation=360)
plt.xticks(ticks=range(0, len(labels), n), labels=labels[::n], rotation=360)

plt.tight_layout()
plt.show()
```
